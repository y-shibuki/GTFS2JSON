```python
import geopandas as gpd
import pandas as pd
import sys

df = pd.read_csv("./kanto_GTFS/stops.txt")
df.head()
```

```python
# location typeが0（標柱）のみを指定
df.query("location_type==0", inplace=True)

df["coord"] = df[["stop_lon", "stop_lat"]].values.tolist()
df[["stop_id", "stop_name", "coord"]].head()
```

```python
sys.setrecursionlimit(10000)
df[["stop_id", "stop_name", "coord"]].to_json("stops.json", orient="records", force_ascii=False)
df[["stop_id", "stop_name", "coord"]].to_csv("./kanto_GTFS/stops_.csv", encoding="utf-8", index=False)
```

```python
df = pd.read_csv("./kanto_GTFS/stops.txt")
# location typeが1（停留所）のみを指定
df.query("location_type==1", inplace=True)
gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.stop_lon, df.stop_lat)).to_file("stops_停留所.geojson")
```
